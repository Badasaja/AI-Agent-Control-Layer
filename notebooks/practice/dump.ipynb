```python
from collections import deque
```

```python
my_queue = deque()
```

```python
my_queue.append("영상_1초_구간")
my_queue.append("영상_2초_구간")
my_queue.append("영상_3초_구간")
```

```python
my_queue
```

```python
while my_queue:
    # 가장 왼쪽(앞) 데이터 꺼내기
    item = my_queue.popleft() 
    print(f"재생 중: {item}")
    print(f"남은 큐: {my_queue}")
```
